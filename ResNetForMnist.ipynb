```python
from tensorflow.keras.datasets import fashion_mnist

(x_train,t_train),(x_test,t_test) = fashion_mnist.load_data()
```

```python
import numpy as np

t_train = np.identity(10)[t_train]
t_test = np.identity(10)[t_test]
x_train = x_train/255
x_test = x_test/255
x_train = x_train[:,:,:,np.newaxis]
x_test = x_test[:,:,:,np.newaxis]
from sklearn.model_selection import train_test_split

x_train,x_val,t_train,t_val = train_test_split(x_train,t_train,test_size = 0.2)
```

```python
from ResNetForIllustration import *
```

```python
trainer = MnistTrainer(
    input_shape = (28,28,1),
    output_dim = 10
)
```

```python
trainer.train(
    x_train,
    t_train,
    x_val,
    t_val,
    epochs = 30,
    batch_size = 300
)
```

```python
trainer.evaluate(x_test,t_test)
```

```python
trainer.train(
    x_train,
    t_train,
    x_val,
    t_val,
    epochs = 2,
    batch_size = 1024
)
```

```python
trainer.model.save_weights('./logs/resnet_for_fashion_mnist')
```

```python
from sklearn.datasets import load_digits
from ResNetForIllustration import *
```

```python
digits = load_digits()
```

```python
data = digits.data
target = digits.target

import numpy as np
from sklearn.model_selection import train_test_split
data.shape
```

```python
data = data.reshape((-1,8,8,1))/16

target = np.identity(10)[target]

x_train,x_test,t_train,t_test = train_test_split(data,target,test_size = 0.2)
x_train,x_val,t_train,t_val = train_test_split(x_train,t_train,test_size = 0.2)
```

```python
x_train.shape
```

```python
trainer = MnistTrainer(
    input_shape = (8,8,1),
    output_dim = 10
)
```

```python
trainer.train(
    x_train,
    t_train,
    x_val,
    t_val,
    epochs = 50,
    batch_size = 128,
    early_stopping = True
)
```
